## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from datetime import datetime

today = datetime.strftime(datetime.now(), "%Y/%m/%d")

browser = webdriver.Chrome(executable_path = "../chromedriver")

browser.get("https://www.ettoday.net/news/news-list.htm")

for i in range(50):
    time.sleep(2)
    
    browser.execute_script("window.scrollTo(0, 50000);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for post in soup.find(class_ = "part_list_2").find_all('h3'):
    post_time = datetime.strptime(post.span.string, "%Y/%m/%d %H:%M")
    
    if (datetime.strftime(post_time, "%Y/%m/%d") == today):
        print(post.span.string, post.em.string, post.a.string)
    
browser.quit()
```

## 2. 取出現在時間兩小時內的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from datetime import datetime, timedelta

pre_two_hours_time = datetime.strftime(datetime.now() - timedelta(hours = 2), "%Y/%m/%d %H:%M")

browser = webdriver.Chrome(executable_path = "../chromedriver")

browser.get("https://www.ettoday.net/news/news-list.htm")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for post in soup.find(class_ = "part_list_2").find_all('h3'):
    post_time = datetime.strptime(post.span.string, "%Y/%m/%d %H:%M")
    
    if (post_time > datetime.strptime(pre_two_hours_time, "%Y/%m/%d %H:%M")):
        print(post.span.string, post.em.string, post.a.string)
    
browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import time
from datetime import datetime, timedelta, date

three_day_before = datetime.now() - timedelta(days = 3)

pre_two_hours_time = datetime.strftime(datetime.now() - timedelta(hours = 2), "%Y/%m/%d %H:%M")

browser = webdriver.Chrome(executable_path = "../chromedriver")

browser.get("https://www.ettoday.net/news/news-list.htm")

selectYear = Select(browser.find_element_by_id("selY"))
selectYear.select_by_value(str(three_day_before.year))

selectMonth = Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value(str(three_day_before.month))

selectDay = Select(browser.find_element_by_id("selD"))
selectDay.select_by_value(str(three_day_before.day))

browser.find_element_by_id("button").click()

for i in range(50):
    time.sleep(2)
    
    browser.execute_script("window.scrollTo(0, 50000);")
    
target_time_start = datetime(three_day_before.year, three_day_before.month, three_day_before.day, 0, 0, 0) + timedelta(hours = 15)
target_time_end = datetime(three_day_before.year, three_day_before.month, three_day_before.day, 0, 0, 0) + timedelta(hours = 17)

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for post in soup.find(class_ = "part_list_2").find_all('h3'):
    post_time = datetime.strptime(post.span.string, "%Y/%m/%d %H:%M")
    
    if (target_time_start <= post_time <= target_time_end):
        print(post.span.string, post.em.string, post.a.string)
    
browser.quit()
```
