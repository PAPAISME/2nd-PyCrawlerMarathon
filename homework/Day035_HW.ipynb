```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
def assign_to_dict(dict_object, key, value):
    dict_object[key] = value.replace("\n", "").replace("\t", "")

def ptt_show(url):
    rg = requests.get(url)
    
    soup = BeautifulSoup(rg.text, "html5lib")

    for recent_post in soup.find_all("div", class_ = "r-ent"):
        temp_dict = {}
        
        for data in recent_post.find_all("div"):
            if ("author" in data.get("class")):
                assign_to_dict(temp_dict, "作者", data.text)
            
            if ("title" in data.get("class")):
                assign_to_dict(temp_dict, "標題", data.text)
                
            if ("date" in data.get("class")):
                assign_to_dict(temp_dict, "時間", data.text)
                
        for key, value in temp_dict.items():
            print (key, " : ", value)
            
        print()
```

```python
import requests
import time

urls = ["https://www.ptt.cc/bbs/NBA/index.html", 
        "https://www.ptt.cc/bbs/LoL/index.html", 
        "https://www.ptt.cc/bbs/NSwitch/index.html",
        "https://www.ptt.cc/bbs/Gossiping/index.html",
        "https://www.ptt.cc/bbs/C_Chat/index.html"]

startTime = time.time()

for url in urls:
    ptt_show(url)

finishTime = time.time()
print(finishTime - startTime) # 正常情況的爬蟲所需時間
```

```python
import requests
import _thread
import time

urls = ["https://www.ptt.cc/bbs/NBA/index.html", 
        "https://www.ptt.cc/bbs/LoL/index.html", 
        "https://www.ptt.cc/bbs/NSwitch/index.html",
        "https://www.ptt.cc/bbs/Gossiping/index.html",
        "https://www.ptt.cc/bbs/C_Chat/index.html"]

startTime = time.time()

for url in urls:
     _thread.start_new_thread(ptt_show, (url, ))

finishTime = time.time()

print(finishTime - startTime) # 利用 thread 的爬蟲所需時間
```
